# **Створити Subplot із 4 графіків сінусоїди з різними частотами**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
t = np.arange(0, 2, 0.01)

s1 = np.sin(2 * np.pi * t)
s2 = np.sin(4 * np.pi * t)
s3 = np.sin(8 * np.pi * t)
s4 = np.sin(16 * np.pi * t)
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(t, s1, c='r',linewidth=4.0)
ax1.set(xlabel='x',  title='2π*t')
ax1.grid()

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(t, s2, c='orange',linewidth=4.0)
ax2.set(xlabel='x',  title='4π*t')
ax2.grid()

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(t, s3, c='y',linewidth=3.0)
ax3.set(xlabel='x',  title='8π*t')
ax3.grid()

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(t, s4, c='g',linewidth=2.0)
ax4.set(xlabel='x', title='16π*t')
ax4.grid()

plt.show()
```
